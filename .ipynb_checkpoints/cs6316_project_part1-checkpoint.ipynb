### **For Dataset 1:**

#### We need to evaluate our model for Dataset 1. To do this, we'll split our dataset into two parts: a **training set**, which will be used to fit the model, and a **testing set**, which will be used to evaluate the model at the end.

```
# Split the dataset 1
x1 = data1.drop(data1.columns[-1],axis = 1)  # Featured data
x1_n = MinMaxScaler().fit(x1).transform(x1)  # Normalized featured data between 0 to 1
y1 = labels1                                 # Class labels

# Convert numpy array x1_n to dataframe
x1_n = pd.DataFrame(x1_n) 
print("Normalized Featured Data 1 (x1_n)\n{}".format(x1_n.head()))

# Set 30% of the dataset 1 at random for unnormlized testing data
x_train1,x_test1,y_train1,y_test1 = train_test_split(x1,y1,test_size=0.3,
                                                     random_state=0)

# Set 30% of the dataset 1 at random for normalized testing data
x_train1_n,x_test1_n,y_train1_n,y_test1_n = train_test_split(x1_n,y1,test_size=0.3,
                                                     random_state=0)

# Print shapes of trainning set and testing set of dataset 1
print("x_train1's shape: (%d, %d)" % x_train1.shape)
print("y_train1's shape: (%d,)" % y_train1.shape)
print("x_test1's shape: (%d, %d)" % x_test1.shape)
print("y_test1's shape: (%d,)" % y_test1.shape)
```

#### **a. Logistic Regression (LR)**

```
# Create Logistic Regression model
lr1 = LogisticRegression(solver='liblinear',random_state=0)
```

##### **For Unnormalized Dataset 1 (LR)**

```
# Create Logistic Regression model with fitted training data
lr1_fit = lr1.fit(x_train1, y_train1)

# Get actual predictions, based on the probability
pred_lr1 = lr1_fit.predict(x_test1)

# Get accuracy for dataset 1 using Logistic Regression
acc_lr1 = lr1_fit.score(x_test1,y_test1)
print("Accuracy for data1 (LR): %0.3f%%" % (acc_lr1*100))

# Get confusion matrix for dataset1 using Logistic Regression
cm_lr1 = confusion_matrix(y_test1, pred_lr1, labels=[0,1])
fig,ax = plot_confusion_matrix(conf_mat=cm_lr1,
                               show_absolute=True,
                               show_normed=True,
                               colorbar=True)
plt.title("\nComfusion Matrix for Dataset 1 Using Logistic Regression")
plt.show()
```

```
# More comprehensive report on the classification for dataset 1
target_names = ['class label 0', 'class label 1']
print("Classification Report for Data1 Using Logisitic Regression\n")
print("SEE the row of macro avg for PRECISION, RECALL, and F1!!!\n")
print(classification_report(y_test1,pred_lr1,target_names=target_names,digits=3))
```

```
# Find AUC of dataset1 
fpr_lr1, tpr_lr1, thre_lr1 = metrics.roc_curve(y_test1, pred_lr1, pos_label=1)
auc_lr1 = metrics.auc(fpr_lr1, tpr_lr1)
print("AUC using Logistic Regression for data1: %0.3f%%" % (auc_lr1 * 100))

# Plot ROC
plt.plot(fpr_lr1,tpr_lr1,label="data 1, auc="+str(auc_lr1))
plt.title("\nROC of Dataset1 Using Logistic Regression")
plt.show()
```

##### **For Normalized Dataset 1 (LR)**

```
# Create Logistic Regression model with fitted normliazed training data
lr1_fit_n = lr1.fit(x_train1_n, y_train1_n)

# Get actual predictions, based on the probability
pred_lr1_n = lr1_fit_n.predict(x_test1_n)

# Get accuracy for dataset 1 using Logistic Regression
acc_lr1_n = lr1_fit_n.score(x_test1_n,y_test1_n)
print("Accuracy for Normalized data1 (LR): %0.3f%%" % (acc_lr1_n*100))

# Get confusion matrix for normalized dataset1 using Logistic Regression
cm_lr1_n = confusion_matrix(y_test1_n, pred_lr1_n, labels=[0,1])
fig,ax = plot_confusion_matrix(conf_mat=cm_lr1_n,
                               show_absolute=True,
                               show_normed=True,
                               colorbar=True)
plt.title("\nComfusion Matrix for Normalized Dataset 1 Using Logistic Regression")
plt.show()
```

## **I.iv** Adopt **10-fold** **Cross Validation (CV)** to evaluate the performance of all methods on the provided two datasets in terms of **Accuracy**, **Precision**, **Recall**, **F-1 measure**, and **AUC (area under the curve)**.

```
# Prepare the 10-fold cross-validation procedure
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

```
# Citation: https://machinelearningmastery.com/repeated-k-fold-cross-validation-with-python/

# Evaluate a model with a given number of repeats
def evaluate_model(x_test, y_test, rep, K, model):
	# Prepare the cross-validation procedure
	cv = RepeatedKFold(n_splits=K, n_repeats=rep, random_state=1)
	# Evaluate accuracy of model
	acc = cross_val_score(model, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
	return acc
```
